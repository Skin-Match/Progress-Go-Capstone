```python
def check_compatibility(data, product_name, skin_type):
    # Filter baris yang mengandung 'product_name'
    product_rows = data[data['Name'] == product_name]

    # Jika tidak ada produk yang ditemukan, kembalikan 'Tidak diperiksa'
    if product_rows.empty:
        return 'Tidak diperiksa'

    # Periksa kecocokan untuk jenis kulit tertentu
    for _, row in product_rows.iterrows():
        if skin_type == 'normal' and row['Normal'] == 1:
            return 'Cocok!'
        elif skin_type == 'dry' and row['Dry'] == 1:
            return 'Cocok!'
        elif skin_type == 'combination' and row['Combination'] == 1:
            return 'Cocok!'
        elif skin_type == 'sensitive' and row['Sensitive'] == 1:
            return 'Cocok!'
        elif skin_type == 'oily' and row['Oily'] == 1:
            return 'Cocok!'

    # Jika tidak ada kecocokan ditemukan
    return 'Tidak cocok!'

# Tentukan produk dan jenis kulit yang ingin diperiksa
product_name_to_check = 'R.N.A. POWER Face Cream'
skin_type_to_check = 'normal'

# Periksa kecocokan
result = check_compatibility(data, product_name_to_check, skin_type_to_check)

# Output hasil
print(result)
```
